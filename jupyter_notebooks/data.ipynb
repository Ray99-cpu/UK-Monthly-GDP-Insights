```python
df = pd.read_csv("../jupyter_notebooks/data/clean_data/richest_countries.csv")
```

```python
richest_df = df.dropna()# richest_countries is variable name richest_df
```

```python
df_2 = pd.read_csv("../jupyter_notebooks/data/clean_data/corruption.csv")
```

```python
corrupt_df = df_2.dropna()# corruption is variable name corrupt_df
```

```python
wealth_and_corrupt_df = pd.merge(corrupt_df, richest_df, on="country", how="inner")# wealth_and_corrupt_df is variable name wealth_and_corrupt_df
```

```python
wealth_and_corrupt_df.to_csv("../jupyter_notebooks/data/clean_data/wealth_and_corrupt.csv")# wealth is variable name wealth_and_corrupt_df
```

```python
print(wealth_and_corrupt_df.head())
```

```python
selected_countries = ["United States", "Canada", "Germany", "France", "United Kingdom", "Switzerland"]
filtered = wealth_and_corrupt_df[wealth_and_corrupt_df["country"].isin(selected_countries)]
corruption_means = filtered.groupby("country")["corruption_index"].mean()

```

```python
corruption_means.plot(kind="bar", figsize=(10, 6), color='skyblue')
plt.title("Average Corruption Index by Country")
plt.ylabel("Corruption Index")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
corruption_means.plot(kind="pie", figsize=(10, 6), autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral', 'lightskyblue', 'lightpink', 'lightyellow'])
```
